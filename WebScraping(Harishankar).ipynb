## scrapping website

```python
url="https://www.flipkart.com/search?q=camera&sid=jek%2Cp31%2Ctrv&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_2_2_na_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_2_2_na_na_na&as-pos=2&as-type=RECENT&suggestionId=camera%7CDSLR+%26+Mirrorless&requestId=8070e583-4462-4c00-8232-90e701fb9509&as-searchtext=ca"
```

## website response

```python
res=requests.get(url)
```

```python
res
```

## parsing

```python
soup=BeautifulSoup(res.text)
```

```python
soup
```

## camera names

```python
camera=soup.select("._4rR01T")
```

```python
camera
```

```python
price=soup.select("._1_WHN1")
```

```python
price
```

```python
warranty=soup.select("._1AtVbE+ .col-12-12 .rgWa7D:nth-child(5) , .rgWa7D:nth-child(6)")
```

```python
warranty
```

```python
camera_title=[]
```

```python
for item in camera:
    camera_title.append(item.text)
```

```python
camera_title
```

```python
price_title=[]
```

```python
for item in price:
    price_title.append(item.text)
```

```python
price_title
```

```python
warranty_title=[]
```

```python
for item in warranty:
    warranty_title.append(item.text)
```

```python
warranty_title
```

## Camera DataFrame

```python
df=np.DataFrame(camera_title,columns=["camera_title"])
```

```python
df
```

```python
df['price']=price
```

```python
df
```

```python
df['warranty']=warranty
```

```python
df
```

```python
df.to_csv('WS.csv')
```
